### Learning Matplotlib

##### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Matplotlib Pyplot Documentation: (https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.html)
```

##### Basic Graph

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x, y)

plt.title('Basic Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, label='2x', color='r', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='b')

# Setting titles and labels
plt.title('Style a Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')

# X and Y axis Int points
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

# A matplotlib legend is a small box that describes the various elements of the plot in terms of lines and markers that have distinct styles or colors
plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Use Shorthand notation
plt.plot(x, y, 'r^--', label='2x')

# doc: (https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

# Setting titles and labels
plt.title('Style a Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')

# X and Y axis tickmarks points(Scale fo the graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

# A matplotlib legend is a small box that describes the various elements of the plot in terms of lines and markers that have distinct styles or colors
plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize graph
plt.figure(figsize=(5,3), dpi=100) # dpi = pixel per inch

plt.plot(x, y, 'r^--', label='2x')

## Line Number Two
x2 = np.arange(0, 4.5, 0.5)
print(f"x2 = {x2}")

# plotting x2 and y=(x2^2)
plt.plot(x2, x2**2, 'b', label='x^2')

plt.title('Style a Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

plt.legend()

# save your graph (when saving try keeping dpi to 300 if possible)
plt.savefig('helpers/mygraph.png', dpi=300)

plt.show()
```

```python
x2 = np.arange(0,4.5,0.5)
print(f"x2 = {x2}")

# Plot 0 to 5-th index as solid line
plt.plot(x2[:6], x2[:6]**2, 'b', label='x^2')
# Plot rest of the graph as dashed line
plt.plot(x2[5:], x2[5:]**2, 'b--',)

plt.title('Style a Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X axis')
plt.ylabel('Y axis (X^2)')

plt.xticks([0,1,2,3,4])

plt.legend()

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# Or,
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

## Real World Examples

### Load Gas Price Data

```python
gas = pd.read_csv('helpers/gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas Prices over time (in USD)")
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

plt.xticks(gas.Year[::3]) # take all years, step size 3

plt.legend()
plt.show()
```

##### Another way to plot

```python
plt.figure(figsize=(8,5))
plt.title("Gas Prices over time (in USD)", fontdict={'fontweight': 'bold', 'fontsize': 18})
plt.xlabel('Year')
plt.ylabel('US Dollars')

for country in gas:
    if(country != 'Year'):
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3]) # take all years, step size 3

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig('helpers/Gas_price_figure.png', dpi=300)
plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('helpers/fifa_data.csv')
fifa.head(5)
```

### Histograms

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#22bd62')
plt.xticks(bins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player SKills in FIFA 2018')
plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0]

labels = ['Left', 'Right']
colors = ['#0afaf2', '#9f2ffa']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f%%')

plt.title('Foot Preference of Fifa Players')

plt.show()
```
